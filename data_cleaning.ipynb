**Cleaning Empty Cells**

```python
# Creating another copy of the dataframe

import pandas as pd
df = pd.read_csv('data.csv',index_col=0)
new_df = df.dropna()
print(new_df.to_string())
```

```python
# If you want to change the original DataFrame, use the inplace = True argument

import pandas as pd
df = pd.read_csv('data.csv', index_col=0)
df.dropna(inplace = True)
print(df.to_string())
```

```python
# Replacing Empty Values

import pandas as pd
df = pd.read_csv('data.csv')
df.fillna(130, inplace = True)
df
```

```python
# Replacing Only For Specified Columns

import pandas as pd
df = pd.read_csv('data.csv', index_col=0)
df["Calories"].fillna(130, inplace = True)
df
```

```python
# Replacing Using Mean, Median, or Mode

# mean
import pandas as pd
df = pd.read_csv('data.csv', index_col=0)
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
df
```

```python
# Median
import pandas as pd
df = pd.read_csv('data.csv', index_col=0)
x = df["Calories"].median()
df["Calories"].fillna(x, inplace = True)
df
```

```python
# Mode
import pandas as pd
df = pd.read_csv('data.csv', index_col=0)
x = df["Calories"].mode()[0]
df["Calories"].fillna(x, inplace = True)
df
```

**Fixing Wrong Data**

```python

df.loc[7, 'Duration'] = 45 # Instead of 450 which is impossible
df
```

```python
# Deleting rows where "Duration" is higher than 120
for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120
```

**Removing Duplicates**

```python
# To discover duplicates, we can use the duplicated() method.
# The duplicated() method returns a Boolean values for each row:

print(df.duplicated())
```

```python
# Removing Duplicates

df.drop_duplicates(inplace = True)
print(df.duplicated())
```
